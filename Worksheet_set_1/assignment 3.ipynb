1. Write a python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.
For e.g. If user input is ‘guitar’. Then search for guitars.


```python
#lets now import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException   #Importing Exceptions

import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url='https://www.amazon.in/'
driver.get(url)

search_field=driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]')
search_field

#write on search bar
#[guitar,sunglasses,laptops]
search_field.send_keys(input("Search Results for :-"))
search_btn=driver.find_element_by_xpath('//*[@id="nav-search-submit-button"]')

search_btn.click()
```

2. In the above question, now scrape the following details of each product listed in first 3 pages 
of your search results and save it in a data frame and csv. In case if any product vertical has 
less than 3 pages in search results then scrape all the products available under that product 
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of 
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" 
and “Product URL”. In case, if any of the details are missing for any of the product then 
replace it by “-“

```python
Url_links=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')


Urls=[]

for i in Url_links:
    Urls.append(i.get_attribute("href"))
    
len(Urls)
```

```python
#defining variables
Name_of_the_Product=[]
Ratings=[]
NO_Ratings=[]
Prices=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]
Other_Details=[]
```

```python
for i in Urls:
    driver.get(i)
    time.sleep(3)    
    
    #fetching Name_of_the_Product
    try:
        Name_of_the_product=driver.find_element_by_xpath('//*[@id="productTitle"]')
        Name_of_the_Product.append(Name_of_the_product.text.replace("/n","new line"))
    except:
        Name_of_the_Product.append('-')


        
    try:
        Rating=driver.find_element_by_xpath('//*[@id="reviewsMedley"]/div/div[1]/div[2]/div[1]/div/div[2]/div/span/span')
        Ratings.append(Rating.text.replace("/n","new line"))
    except:
        Ratings.append('-')
    
    
    try:
        NO_Rating=driver.find_element_by_xpath('//*[@id="reviewsMedley"]/div/div[1]/div[2]/div[2]')
        NO_Ratings.append(NO_Rating.text.replace("/n","new line"))
    except:
        NO_Ratings.append('-')
        
    try:
        Price=driver.find_element_by_xpath('//*[@id="priceblock_ourprice"]')
        Prices.append(Price.text.replace("/n","new line"))
    except:
        Prices.append('-')
        
        
        
    # fetching Expected_Delivery
    try:
        Expected_delivery=driver.find_element_by_xpath('//*[@id="ddmDeliveryMessage"]/b')
        Expected_Delivery.append(Expected_delivery.text.replace("/n","new line"))
    except:
        Expected_Delivery.append('-')
               
        
    
    #fetching Return_Exchange
    try:
        Return_exchange=driver.find_element_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        Return_Exchange.append(Return_exchange.text.replace("/n","new line"))
    except:
        Return_Exchange.append('-')
        
        
    #fetching Availability
    try:
        availability=driver.find_element_by_xpath('//*[@id="availability"]/span')
        Availability.append(availability.text.replace("/n","new line"))
    except:
        Availability.append('-')
        
     # fetching Other_Details   
    try:
        Other_Detail=driver.find_element_by_xpath('//*[@id="feature-bullets"]/ul/li[2]/span')
        Other_Details.append(Other_Detail.text.replace("/n","new line"))
    except:
        Other_Details.append('-')
        
        
        
print(len(Name_of_the_Product))

print(len(Ratings))
print(len(NO_Ratings))
print(len(Prices))
print(len(Return_Exchange))
print(len(Expected_Delivery))
print(len(Availability))
print(len(Other_Details))        
```

```python
df1=pd.DataFrame({})
df1['Name of the Product']=Name_of_the_Product
df1['Ratings']=Ratings
df1['NO of Ratings']=NO_Ratings
df1['Return Exchange']=Return_Exchange
df1['Expected Delivery']=Expected_Delivery
df1['Availability']=Availability
df1['Other Details']=Other_Details
df1['Product URL']=Urls
df1
```

```python
df1.to_csv('Laptops.csv')
```

# 8. Write a python program to scrape the details for all billionaires from www.forbes.com. 
Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, 
“Industry”.


```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url='https://www.forbes.com/billionaires/'
driver.get(url)
```

```python
Names=[]
Net_Worth=[]
Rank=[]
Age=[]
Country=[]
Source=[]
Category=[]


for i in range(1,14):
    
    rank=driver.find_elements_by_xpath('//div[@class="rank"]')
    for i in rank:
        Rank.append(i.text.replace("\n",""))
    name=driver.find_elements_by_xpath('//div[@class="personName"]')
    for i in name:
        Names.append(i.text.replace("\n",""))
        
    net_worth=driver.find_elements_by_xpath('//div[@class="netWorth"]')
    for i in net_worth:
        Net_Worth.append(i.text.replace("\n",""))

    age=driver.find_elements_by_xpath('//div[@class="age"]')
    for i in age:
        Age.append(i.text.replace("\n",""))
        
    countryOfCitizenship=driver.find_elements_by_xpath('//div[@class="countryOfCitizenship"]')
    for i in countryOfCitizenship:
        Country.append(i.text.replace("\n",""))
    
    
    source=driver.find_elements_by_xpath('//div[@class="source"]')
    for i in source:
        Source.append(i.text.replace("\n",""))
    
    category=driver.find_elements_by_xpath('//div[@class="category"]')
    
    for i in category:
        Category.append(i.text.replace("\n",""))
    
    

    nxt_button=driver.find_element_by_xpath('//div[@class="next-page"]')
    nxt_button.click()
   
```

```python
df1=pd.DataFrame({})
df1['Rank']=Rank
df1['Names']=Names
df1['Net_Worth']=Net_Worth
df1['Age']=Age
df1['Country']=Country
df1['Source']=Source
df1['Industry']=Category

df1
```

```python
df1.to_csv('Billionairs_data.csv')
```

# 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment 
was posted from any YouTube Video.


```python
import time
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
wait = WebDriverWait(driver,10)
driver=webdriver.Chrome("chromedriver.exe")

```

```python
url='https://www.youtube.com/watch?v=PgCliOxl41o'
driver.get(url)
```

```python
data=[]


wait = WebDriverWait(driver,10)
    

for item in range(300): 
    wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
    time.sleep(10)

for comment in wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#content-text"))):
    data.append(comment.text)
```

```python
len(data)
```

```python
likes=driver.find_elements_by_xpath("//span[@class='style-scope ytd-comment-action-buttons-renderer']")
upvote=[]
for i in likes:
    upvote.append(i.text)
```

```python
Upvotes=[]
for i in range(1,len(upvote),2):
    try:
        Upvotes.append(upvote[i].strip())
    except:
        Upvotes.append("-")
len(Upvotes) 
```

```python
time=driver.find_elements_by_xpath("//yt-formatted-string[@class='published-time-text above-comment style-scope ytd-comment-renderer']")
times=[]
for i in time:
    times.append(i.text)
```

```python
len(times)
```

```python
df1=pd.DataFrame({})
df1['Comments']=data
df1['Upvotes']=Upvotes
df1['Time of Comment']=times
df1
```

```python
df1.to_csv('comments.csv')
```
